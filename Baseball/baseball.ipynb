```python
# https://www.kaggle.com/open-source-sports/baseball-databank
import pandas as pd
import numpy as np
import os
```

```python
b_df = pd.read_csv('BattingPost.csv')
print(b_df.columns)
b_df.head()
```

```python
weight_cols = ['G', 'AB', 'R', 'H', '2B', '3B', 'HR', 'RBI', 'SB',
               'CS', 'BB', 'SO', 'IBB', 'HBP', 'SH', 'SF', 'GIDP']
```

```python
for year, df in b_df.groupby('yearID'):
    print(year)
    df['id'] = df['round'] + '/' + df['lgID'] + '/' + df['teamID'].astype('str') + '/' + df['playerID']
#         display(df)
    for col in weight_cols:
        copy = df.loc[:, ['id', col]].dropna(axis=0).copy()
        copy.columns = ['id', 'weight']
        copy['weight'] = copy['weight'].abs()  # We can't have negative weights
        copy['weight'] = copy['weight'].replace({0:np.nan})
        copy.dropna(axis=0, how='any', inplace=True)

        dir_name = 'baseball/batting-' + col + '/'
        os.makedirs(dir_name, exist_ok=True)
        if copy.shape[0] > 0:  # Check if not empty
            copy.to_csv(dir_name + str(year) + '-'  + str(copy.shape[0]) + '.csv', index=False)
```
