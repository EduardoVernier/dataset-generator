```python
# https://www.kaggle.com/open-source-sports/mens-professional-basketball#basketball_players.csv
import pandas as pd
import numpy as np
import os
```

```python
player_df = pd.read_csv('basketball_players.csv')
```

```python
player_df[player_df['playerID'] == 'abramjo01']
```

```python
weight_cols = ['GP', 'GS', 'minutes', 'points', 'oRebounds', 'dRebounds', 'rebounds', 'assists', 'steals',
       'blocks', 'turnovers', 'PF', 'fgAttempted', 'fgMade', 'ftAttempted',
       'ftMade', 'threeAttempted', 'threeMade', 'PostGP', 'PostGS',
       'PostMinutes', 'PostPoints', 'PostoRebounds', 'PostdRebounds',
       'PostRebounds', 'PostAssists', 'PostSteals', 'PostBlocks',
       'PostTurnovers', 'PostPF', 'PostfgAttempted', 'PostfgMade',
       'PostftAttempted', 'PostftMade', 'PostthreeAttempted', 'PostthreeMade']
```

```python
for year, df in player_df.groupby('year'):
    print(year)
    df['id'] = df['lgID'] + '/' + df['tmID'] + '/' + df['stint'].astype('str') + '/' + df['playerID']
    for col in weight_cols:
        copy = df.loc[:, ['id', col]].dropna(axis=0).copy()
        copy.columns = ['id', 'weight']
        copy['weight'] = copy['weight'].abs()  # We can't have negative weights
        copy['weight'] = copy['weight'].replace({0:np.nan})
        copy.dropna(axis=0, how='any', inplace=True)

        dir_name = 'basketball/' + col + '/'
        os.makedirs(dir_name, exist_ok=True)
        if copy.shape[0] > 0:  # Check if not empty
            copy.to_csv(dir_name + str(year) + '-'  + str(copy.shape[0]) + '.csv', index=False)
```
