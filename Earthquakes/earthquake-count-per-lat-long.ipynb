```python
import pandas as pd
from IPython.display import display
```

```python
df = pd.read_csv('earthquakes.csv')
```

```python
df.info()
```

```python
df['time'] = pd.to_datetime(df['time'],infer_datetime_format=True)
df['time'] = df['time'].dt.date
df['time'] = df['time'].astype(str)
```

```python
df['latitude'] = df['latitude'].astype(int).astype(str)
df['longitude'] = df['longitude'].astype(int).astype(str)
df['coords'] = (df['latitude']).str.cat(df['longitude'], sep='^')
```

```python
df = df.loc[:, ['time', 'coords', 'mag']]
```

```python
df
```

```python
for name, group in df.groupby('time'):
    temp = group.groupby('coords').count()
    temp['id'] = [name + '/' + i for i in temp.index]
    temp['weight'] = temp['mag']
    temp.to_csv('earthquake-count-per-lat-long/' + name + '-earthquakes.csv', columns=['id','weight'], index=False)
    
```
