# Scrape the first 10 pages of gitmostwanted.com

```python
from bs4 import BeautifulSoup
import requests
```

```python
request_list = ['http://gitmostwanted.com/top/wanted/solid/' + str(i) for i in range(1, 11)]
request_list
```

```python
for request in request_list:
    page = requests.get(request)
    soup = BeautifulSoup(page.content, 'html.parser')
    urls = soup.find_all('li', attrs={'class': 'promising'})
    for url in urls:
        # Filter shell and markdown repos
        if url.find('small') and url.find('small').text != 'Shell':     
            print(url['data-url'])
```
